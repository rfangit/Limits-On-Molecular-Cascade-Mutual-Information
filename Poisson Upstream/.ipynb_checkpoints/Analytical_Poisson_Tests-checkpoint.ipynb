# Steady State Distribution Tests

## Analytical Time-Dependent Solution vs Numerical Integration

An example of numerically integrating the binary markovian system to obtain the steady state distribution, as well as a comparison with the analytical time-dependent distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

figure_file = "binary_markov_ode_solution.png"

# Define parameters
k_on = 1.0   # Transition rate to x_on
k_off = 0.5  # Transition rate to x_off

# Initial conditions
P0_on = 1.0  # Probability of being in state x_on at t=0
P0_off = 0.0 # Probability of being in state x_off at t=0

# Time range
t_span = (0, 10)
t_eval = np.linspace(t_span[0], t_span[1], 100)

# System of ODEs
def two_state_system_dt(t, p, k_off, k_on):
    dp0dt = -k_on * p[0] + k_off * p[1]
    dp1dt = k_on * p[0] - k_off * p[1]
    return np.array([dp0dt, dp1dt])

# Solve the ODEs numerically
sol = solve_ivp(two_state_system_dt, t_span, [P0_off, P0_on], t_eval=t_eval, args=(k_off, k_on))

# Compute the analytical solution
P_eq_on = k_on / (k_on + k_off)
P_eq_off = k_off / (k_on + k_off)
C2 = (P0_on - P_eq_on)
P_analytic_on = P_eq_on + C2 * np.exp(-(k_on + k_off) * t_eval)
P_analytic_off = P_eq_off - C2 * np.exp(-(k_on + k_off) * t_eval)

# Plot numerical vs analytical solutions
plt.figure(figsize=(8, 5))
plt.plot(t_eval, P_analytic_on, 'b', lw=2, label='Analytical $P(x_{on})$')
plt.plot(t_eval, P_analytic_off, 'r', lw=2, label='Analytical $P(x_{off})$')
plt.plot(t_eval, sol.y[0], 'm--', lw=2, label='Numerical $P(x_{off})$')
plt.plot(t_eval, sol.y[1], 'g--', lw=2, label='Numerical $P(x_{on})$')
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Probability', fontsize = 15)
plt.legend()
plt.title('Comparison of Numerical and Analytical Solutions', fontsize = 15)
plt.grid()
plt.savefig(figure_file)
plt.show()
```

## Analytical Results vs Gillespie Simulation

Instead of numerically integrating the ODEs governing the probability distribution, we run a gillespie simulation and acquire the probability distribution of system states (as measured by the average time spent in each state over the simulation). We compare these results with our analytical solution for the steady state solution.

This serves additionally as verification of our gillespie simulation code.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
k_on = 1.0   # Transition rate to x_on
k_off = 0.5  # Transition rate to x_off
num_reactions = 10000  # Fixed number of reactions to simulate
num_simulations = 10  # Number of Gillespie runs
initialization_steps = 500  # Number of initial steps before tracking
output_file = "gillespie_results.txt"
figure_file = "binary_markov_gillespie_plot.png"

# Function to perform a single Gillespie simulation
def gillespie_two_state(num_reactions, initialization_steps, k_on, k_off):
    state = 1  # Start in x_on state
    t = 0
    last_t = 0
    state_durations = []
    
    for step in range(num_reactions):
        rate_on = k_on * (1 - state)
        rate_off = k_off * state
        total_rate = rate_on + rate_off
        if total_rate == 0:
            break  # No transitions possible
        dt = np.random.exponential(1 / total_rate)
        t += dt
        if step >= initialization_steps:
            state_durations.append([state, t - last_t])
        last_t = t
        if np.random.rand() < (rate_off / total_rate):
            state = 0  # Switch to x_off
        else:
            state = 1  # Switch to x_on
    
    return state_durations

# Run multiple Gillespie simulations and save results
all_simulations = []
for _ in range(num_simulations):
    state_durations = gillespie_two_state(num_reactions, initialization_steps, k_on, k_off)
    all_simulations.extend(state_durations)

np.savetxt(output_file, np.array(all_simulations), fmt='%d %.6f')
print(f"Simulation results saved to {output_file}")

# Function to compute total time spent in each state
def compute_state_times(all_simulations):
    total_time_in_on = np.sum([duration for state, duration in all_simulations if state == 1])
    total_time_in_off = np.sum([duration for state, duration in all_simulations if state == 0])
    total_time = total_time_in_on + total_time_in_off
    return total_time_in_on, total_time_in_off, total_time

# Load results and perform analysis
all_simulations = np.loadtxt(output_file)
total_time_in_on, total_time_in_off, total_time = compute_state_times(all_simulations)

# Compute empirical probabilities based on time spent in each state
P_gillespie_on = total_time_in_on / total_time
P_gillespie_off = total_time_in_off / total_time

# Theoretical equilibrium probabilities
P_eq_on = k_on / (k_on + k_off)
P_eq_off = k_off / (k_on + k_off)

# Display results
print(f"Gillespie Simulated P(X_on) = {P_gillespie_on:.4f}, Theoretical P(X_on) = {P_eq_on:.4f}")
print(f"Gillespie Simulated P(X_off) = {P_gillespie_off:.4f}, Theoretical P(X_off) = {P_eq_off:.4f}")

# Plot histogram of final states
plt.bar(["X_on", "X_off"], [P_gillespie_on, P_gillespie_off], color=['blue', 'red'], alpha=0.7, label='Gillespie')
plt.axhline(P_eq_on, color='blue', linestyle='--', label='Theoretical $P(X_{on})$')
plt.axhline(P_eq_off, color='red', linestyle='--', label='Theoretical $P(X_{off})$')
plt.ylabel("Probability", fontsize = 15)
plt.xticks(fontsize=12)  # Adjust fontsize of X_on and X_off labels
plt.legend()
plt.title("Gillespie Simulation vs. Theoretical Distribution", fontsize = 15)
plt.savefig(figure_file)
plt.show()

print(f"Figure saved to {figure_file}")
```

# Analytical Solutions for Joint Probability Distributions Vs ODE Numerical Integration

There are a number of analytical joint probability distributions to verify.

## Survival Probability

We derived an expression for the probability that given a detection event at time $t = 0$, no new detection events have occured. The underlying system of ODEs is given by

\begin{equation}
\frac{d P(x_{on}, t_l )}{dt} = - (\alpha + k_{off}) P(x_{on}, t_l) + k_{on} P(x_{off}, t_l)
\end{equation}
\begin{equation}
\frac{d P(x_{off}, t_l)}{dt} = - k_{on} P(x_{off}, t_l) + k_{off} P(x_{on}, t_l)
\end{equation}

which we solve numerically and analytically and compare.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

figure_file = "detection_event_binary_markov_ode_solution.png"

# Define parameters
k_on = 1.0     # Transition rate to x_on
k_off = 1.0    # Transition rate to x_off
alpha = 0.2    # Additional rate parameter for x_on state

# Initial conditions
P0_on = 1.0    # Probability of being in state x_on at t=0
P0_off = 0.0   # Probability of being in state x_off at t=0

# Time range
t_span = (0, 50)
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# The system of ODEs describing the evolution of the binary markov system, with detection events
def two_state_system_detection_dt(t, p, k_off, k_on, alpha):
    dp0dt = -k_on * p[0] + k_off * p[1]
    dp1dt = k_on * p[0] - k_off * p[1] - alpha*p[1]
    return np.array([dp0dt, dp1dt])

# Solve the ODEs numerically
sol = solve_ivp(two_state_system_detection_dt, t_span, [P0_off, P0_on], t_eval=t_eval, args=(k_off, k_on, alpha))

# Analytical solution to the system of ODEs with detection events
# Returns a pair of arrays Pon, Poff
# These arrays denote the probability that:
# a new detection event has not occured
# it is currently a time t after the initial detection event
# and the system is in the on or off states
def two_state_analytical_solution(k_on, k_off, alpha, t_eval):
    # Calculate eigenvalues
    a = alpha + k_off + k_on
    D = np.sqrt(a**2 - 4*alpha*k_on)
    lambda1 = (-a + D)/2
    lambda2 = (-a - D)/2
    denominator = lambda2 - lambda1

    # Calculate probabilities
    P_on = ((alpha + k_off + lambda2) * np.exp(lambda1 * t_eval) - 
            (alpha + k_off + lambda1) * np.exp(lambda2 * t_eval)) / denominator
    P_off = -k_off * (np.exp(lambda1 * t_eval) - np.exp(lambda2 * t_eval)) / denominator
    return P_on, P_off

# Compute the analytical solution
P_analytic_on, P_analytic_off = two_state_analytical_solution(k_on, k_off, alpha, t_eval)

# Plot numerical vs analytical solutions
plt.figure(figsize=(10, 6))
plt.plot(t_eval, P_analytic_on, 'b', lw=2, label='Analytical $P(x_{on})$')
plt.plot(t_eval, P_analytic_off, 'r', lw=2, label='Analytical $P(x_{off})$')
plt.plot(t_eval, sol.y[0], 'm--', lw=2, label='Numerical $P(x_{on})$')
plt.plot(t_eval, sol.y[1], 'g--', lw=2, label='Numerical $P(x_{off})$')
plt.xlabel('Time', fontsize=15)
plt.ylabel('Probability', fontsize=15)
plt.xlim((0, 20))
plt.legend(fontsize=12)
plt.title('Survival Probability with  = {0}'.format(alpha), fontsize=15)
plt.grid(True)
plt.savefig(figure_file, bbox_inches='tight', dpi=300)
plt.show()
```

## Distribution of Event Times

Given a detection event occurred at time $t = 0$, there is a probability that the next detection event occurs at time $t$. This is given by the derivative w.r.t to time of our survival probabilities above, which we plot here.

```python
# Analytical distribution of detection event times
# Returns an array, denoting the probability that the next detection event occurs
# at a time t_eval[time_index] after the last detection event
def event_time_dist(k_on, k_off, alpha, t_eval):
    # Calculate eigenvalues
    a = alpha + k_off + k_on
    D = np.sqrt(a**2 - 4*alpha*k_on)
    lambda1 = (-a + D)/2
    lambda2 = (-a - D)/2
    denominator = lambda2 - lambda1
    
    numerator = (alpha + lambda1)*lambda2*np.exp(lambda2*t) - (alpha + lambda2)*lambda1*np.exp(lambda1*t)
    return numerator / denominator

# Time points
t = np.linspace(0, 100, 10000)  # From 0 to 5 time units

# Calculate function values
event_times = event_time_dist(k_on, k_off, alpha, t)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(t, event_times, 'g-', linewidth=2)
#plt.plot(t_eval, sol.y[0], 'm--', lw=2, label='Numerical $P(x_{on})$')

plt.title('Probability Density Function of Event Times')
plt.xlabel('Time (t)')
plt.ylabel('Probability Density f(t)')
plt.savefig("detection_event_times_binary_markov.png", bbox_inches='tight', dpi=300)
plt.grid(True)
plt.show()
```

# Gillespie Simulation and Entropy Results

## Gillespie Simulation

Here, we run a gillespie simulation of our system, and track:

- time between detection events (yielding the earlier detection event time distribution)
- time since the last detection event, at random times (corresponds to the distribution of survival probabilities, but normalized to 1)
- time spent in each state, and the ending reaction of the state (either a state transition, or a detection event)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
k_on = 1.0   # Transition rate to x_on
k_off = 1.0  # Transition rate to x_off
alpha = 0.2
num_reactions = 200000  # Fixed number of reactions to simulate
num_simulations = 10  # Number of Gillespie runs
initialization_steps = 100  # Number of initial steps before tracking
sampling_spacing = 10
output_file = "detection_gillespie_results.txt"
figure_file = "detection_binary_markov_gillespie_plot.png"

# Function to perform a single Gillespie simulation
def gillespie_two_state_detection(num_reactions, initialization_steps, sampling_spacing, k_on, k_off, alpha):
    state = 1  # Start in x_on state
    t = 0
    last_t = 0
    last_detection = 0
    state_durations = []
    detection_times = []
    time_since_detection_samples = []
    
    for step in range(num_reactions):
        rate_on = k_on * (1 - state)
        rate_off = k_off * state
        rate_detection = alpha * state
        total_rate = rate_on + rate_off + rate_detection
        
        dt = np.random.exponential(1 / total_rate)
        # As our simulation runs, we sample its state every sampling_spacing, and grab information about the current state
        # and the time since the last detection event
        if step >= initialization_steps:
            i = (int)(t/sampling_spacing)
            if t < sampling_spacing*(i+1) and t + dt > sampling_spacing*(i+1): ##if we cross a sample threshold
                time_since_last_detection = sampling_spacing*(i+1) - last_detection
                time_since_detection_samples.append([state, time_since_last_detection])
        t += dt
        
        last_state = state
        detection = 0
        which_event_happened = np.random.rand()
        if which_event_happened < (rate_off / total_rate):
            state = 0  # Switch to x_off
        else:
            if which_event_happened < ((rate_off + rate_on) / total_rate):
                state = 1  # Switch to x_on
            else:
                detection = 1 
                ## we also gather a specialized dataset of only the time between detection events
                if step >= initialization_steps and last_detection != 0:
                    detection_times.append([t - last_detection])
                last_detection = t
                
        if step >= initialization_steps:
            state_durations.append([last_state, dt, detection]) ##if it was a detection event, it'll be recorded here
        last_t = t
    
    return state_durations, detection_times, time_since_detection_samples

all_state_durations = []
all_detection_times = []
all_time_since_detection_samples = []

for i in range(num_simulations):
    state_durations, detection_times, time_since_detection_samples = gillespie_two_state_detection(
        num_reactions, initialization_steps, sampling_spacing, k_on, k_off, alpha
    )
    all_state_durations.extend(state_durations)
    all_detection_times.extend(detection_times)
    all_time_since_detection_samples.extend(time_since_detection_samples)

np.savetxt("state_durations.txt", np.array(all_state_durations))
np.savetxt("detection_times.txt", np.array(all_detection_times))
np.savetxt("time_since_detection.txt", np.array(all_time_since_detection_samples))
#print (all_detection_times)

print("Simulation results saved to state_durations.txt, detection_times.txt, and time_since_detection.txt")
```

These results are saved into txt files for analysis.

### Entropy of X

As a simple verification of our simulation's accuracy, we check the steady state distribution of x, and also compute its entropy.

```python
# Compute total time spent in each state with optional row limit
num_rows = min(len(state_durations_data), 1000000)  # Change 1000000 to any limit you want
max_state = int(np.max(state_durations_data[:num_rows, 0])) + 1  # Determine the number of states
state_time_totals = np.zeros(max_state)

for row in state_durations_data[:num_rows]:
    state, time_spent, _ = row
    state_time_totals[int(state)] += time_spent

print("Total time spent in each state (limited to", num_rows, "rows):", state_time_totals)

# Calculate probabilities by normalizing
total_time = np.sum(state_time_totals)
state_probabilities = state_time_totals / total_time

print("\nProbability distribution of states:")
for state, prob in enumerate(state_probabilities):
    print(f"State {state}: {prob:.4f}")

#print (state_probabilities[0])
# Compute entropy (in natural units)
entropy = -np.sum([p * np.log(p) for p in state_probabilities if p > 0])

print(f"\nEntropy of the state distribution: {entropy:.4f} nats")
```

## Distribution of Event Times

We compare the distribution times between detection events from our simulation with our theoretical results, using histograms. We also show our analytical solution gives a superior fit to the distribution of event times than a simple exponential fit, or fitting alpha as the decay exponential.

```python
# Plot histogram of detection times with customizable settings
bin_size = 500  # Adjust the number of bins
x_limit = (0, 30)  # Set x_limit to a value like (0, 100) to restrict the x-axis

plt.hist(detection_times_data, bins=bin_size, density=True, edgecolor='black', alpha=0.7, label="Histogram")
plt.xlabel("Detection Time")
plt.ylabel("Probability")
plt.title("Histogram of Detection Times")

# Fit and plot an exponential distribution overlay
lambda_estimate = 1 / np.mean(detection_times_data)  # Estimate lambda as 1/mean
print ("Fitting to an exponential curve results in the following decay constant ", lambda_estimate)
x_values = np.linspace(x_limit[0], x_limit[1], 1000)
y_values = lambda_estimate * np.exp(-lambda_estimate * x_values)
plt.plot(x_values, y_values, 'r-', label="Exponential Fit")

# Fit the known decay rate alpha
lambda_estimate2 = alpha
y_values2 = lambda_estimate2 * np.exp(-lambda_estimate2 * x_values)
plt.plot(x_values, y_values2, 'b--', label="Alpha Fit")

# Fit the analytical solution
def event_time_dist(k_on, k_off, alpha, t_eval):
    # Calculate eigenvalues
    a = alpha + k_off + k_on
    D = np.sqrt(a**2 - 4*alpha*k_on)
    lambda1 = (-a + D)/2
    lambda2 = (-a - D)/2
    denominator = lambda2 - lambda1
    
    numerator = (alpha + lambda1)*lambda2*np.exp(lambda2*t) - (alpha + lambda2)*lambda1*np.exp(lambda1*t)
    return numerator / denominator

# Time points
t = np.linspace(0, 100, 10000)  # From 0 to 5 time units

# Calculate function values
y = event_time_dist(k_on, k_off, alpha, t_eval)
plt.plot(t, y, 'g-', linewidth=2, label = 'Analytical Solution')
plt.legend(loc='best')

if x_limit:
    plt.xlim(x_limit)

plt.savefig("gillespie_detection_event_times_binary_markov.png", bbox_inches='tight', dpi=300)
plt.show()
```

## Analytical Joint Distribution vs Gillespie Results

In our simulation earlier, we queried the system at regular intervals to collect data on the current state, and the time since the last detection event. We plot this data as a histogram, which we compare to our theoretical predictions for the survival probabilities $P(0, t), P(1, t)$.

```python
# Load your data
time_since_detection_data = np.loadtxt("time_since_detection.txt")

# Get all unique state numbers
unique_states = np.unique(time_since_detection_data[:, 0]).astype(int)
num_states = len(unique_states)
max_state = np.max(unique_states)

# Create a list to hold arrays for each state
# Initialize with empty arrays for all possible states
state_times = [np.array([]) for _ in range(max_state + 1)]

# Separate the data
for state in unique_states:
    # Get all times for current state
    mask = time_since_detection_data[:, 0] == state
    state_times[state] = time_since_detection_data[mask, 1]

# Print summary
print(f"Separated data into {max_state + 1} states")
for state in unique_states:
    print(f"State {state}: {len(state_times[state])} time points")
    if len(state_times[state]) > 0:
        print(f"  First 3 times: {state_times[state][:3]}")
        print(f"  Mean time: {np.mean(state_times[state]):.2f}")
```

We find good agreement between our histograms and renormalized survival probabilities. Note that renormalization was necessary since the survival probabilities differ from the random sampling probabilities (survival is by default 1 at $t = 0$).
